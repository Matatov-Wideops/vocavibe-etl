```python
import matplotlib.pyplot as plt
from src.plotting import Color
import numpy as np 


def plot_pies():
    
    fig, ax = plt.subplots(figsize=(12, 16))

    # labels = ['PD men', 'PD women', 'HC men', 'HC women']
    labels = ['Satisfactory', 'Warning', 'Urgent', 'Lost contact']

    sizes = [80, 12, 6, 2]
    colors = [Color.LIGHT_GREEN, Color.YELLOW, np.array([255, 20, 50])/255, Color.VV]
    explode = [0.03, 0.05, 0.07, 0.09]



    wedges, texts, autotexts = ax.pie(sizes, explode=explode, colors=colors,
                                             autopct=lambda p: f'{round(p * sum(sizes) / 100)}', shadow=True,
                                             startangle=140)
    for autotext in autotexts:
        autotext.set_fontsize(40)
    plt.legend(labels, fontsize=28, loc='upper right')
    plt.show()

plot_pies()
```
